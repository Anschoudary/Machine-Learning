# **1. Definition and Example for Beginners**

# **k-NN Algorithm:**
# The k-NN algorithm is a non-parametric, supervised learning method used for classification and regression.
# It classifies a data point based on the majority class among its 'k' nearest neighbors in the feature space.

# **Example:**
# Imagine you have a scatter plot of data points with two classes (e.g., red and blue).
# To classify a new data point, you would find the 'k' nearest data points to it.
# If the majority of those neighbors are red, the new data point would be classified as red.



```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

# **2. Import Dataset**
# For this example, we'll use the Iris dataset. You can replace this with your own dataset.

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```


# **3. Preprocess Dataset**
# This involves splitting the data into training and testing sets and scaling the features.

```python
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# **4. Visualize Dataset (Optional)**
# You can visualize the data using scatter plots or other visualization techniques. This step is optional but can be helpful for understanding the data.


```python
# For example, to visualize the first two features:
import matplotlib.pyplot as plt
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

# **5. Train Model**

```python
knn = KNeighborsClassifier(n_neighbors=5)  # You can change the value of 'k'
knn.fit(X_train, y_train)
```

# **6. Accuracy Score**

```python
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

# **7. Confusion Matrix**

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

# **8. Check the Model**

```python
def test_model(model, X_test, y_test):
    """
    Tests the given model using the provided test data and prints the accuracy and confusion matrix.

    Args:
      model: The trained machine learning model.
      X_test: The features of the test data.
      y_test: The true labels of the test data.
    """
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy}")
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix:\n{conf_matrix}")

# Example usage (assuming you have a trained knn model):
test_model(knn, X_test, y_test)
```
